The script FAFMIP_binned has binned tendency terms in T-z (and S-z, if that's your flavour) and saved them as .nc files for all FAFMIP experiments available on NCI. I here load in those .nc files and examine the heat budget in T-z space and the T-z streamfunctions for faf-passiveheat and faf-heat experiments.

- I show how some models are weird in their residual mean advective tendency terms (e.g., CAS-ESM2-0).

- I also examine how well the heat budget closes in T-z space for a couple of models.

- I then briefly examine streamfunctions in T-z space for faf-passiveheat and faf-heat where available.

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TwoSlopeNorm
import pandas as pd
import netCDF4 as nc
```

```python
path_fs38 = '/g/data/fs38/publications/CMIP6/FAFMIP/' # For ACCESS-CM2
path_oi10 = '/g/data/oi10/replicas/CMIP6/FAFMIP/'     # For models not ACCESS-CM2
```

```python
def tendency_term_binnned_streamfunction(tendency_term_binned, binned_coordinate):
    '''
    DESCRIPTION:
    Calculate the coordinate-depth streamfunction for the already-binned tendency_term.
    ---
    OPTIONS:
    ---
    tendency_term_binned:  A three-dimensional (year, lev, coordinate_bin) xarray DataArray
    binned_coordinate:     'temperature', 'salinity'
    ---
    '''
    
    cp = 3992.10322329649  # Constant heat capacity
    rho0 = 1035            # Reference density
    
    if binned_coordinate == 'temperature':
    
        tendency_term_streamfunction = - (tendency_term_binned.cumsum('lev') - tendency_term_binned.sum('lev')) / (cp * rho0 * 1e6)
        
    elif binned_coordinate == 'salinity':
        
        tendency_term_streamfunction = - (tendency_term_binned.cumsum('lev') - tendency_term_binned.sum('lev')) / (1e6)
    
    return tendency_term_streamfunction
```

### Looking at residual mean advective tendency terms to see how some models (e.g., CAS, MRI) might be a bit strange (see their streamfunctions too).

```python
def rmadvect_temp_terms(models, faf_expt_name):
    
    variables = []
    
    for model_name in models:
    
        ### Match model_name to modelling_group ###

        if model_name == 'ACCESS-CM2':
            modelling_group = 'CSIRO-ARCCSS'
        elif model_name == 'CAS-ESM2-0':
            modelling_group = 'CAS'
        elif model_name == 'CanESM5':
            modelling_group = 'CCCma'
        elif model_name == 'FGOALS-g3':
            modelling_group = 'CAS'
        elif model_name == 'MIROC6':
            modelling_group = 'MIROC'
        elif model_name == 'MPI-ESM1-2-HR':
            modelling_group = 'MPI-M'
        elif model_name == 'MRI-ESM2-0':
            modelling_group = 'MRI'
        elif model_name == 'CESM2':
            modelling_group = 'NCAR'
        elif model_name == 'GFDL-ESM2M':
            modelling_group = 'NOAA-GFDL'


        ### Define path to FAFMIP experiments ###

        if model_name == 'ACCESS-CM2':
            model_path = path_fs38 + modelling_group + '/' + model_name + '/'
        else:
            model_path = path_oi10 + modelling_group + '/' + model_name + '/'

        expt_path = model_path + faf_expt_name + '/*/'
        
        name = 'rmadvect'

        if model_name == 'ACCESS-CM2':
            # different path ending...!
            
            temperature_type = 'ocontemp'
            variables.append(xr.open_mfdataset(expt_path + 'Oyr/' + temperature_type + name + '/gn/latest/*.nc')[temperature_type + name].groupby('time.year').mean(dim='time', keep_attrs=True).load())
        
        elif model_name == 'MRI-ESM2-0':
            # path has 'gr' instead of 'gn' for monthly variables...!
            
            temperature_type = 'opottemp'
            variables.append(xr.open_mfdataset(expt_path + 'Oyr/' + temperature_type + name + '/gn/*/*.nc')[temperature_type + name].groupby('time.year').mean('time', keep_attrs=True).load())
                            
        
        else:
            temperature_type = 'opottemp'
            variables.append(xr.open_mfdataset(expt_path + 'Oyr/' + temperature_type + name + '/gn/*/*.nc')[temperature_type + name].groupby('time.year').mean('time', keep_attrs=True).load())
            
    return variables
```

```python
model = ['ACCESS-CM2', 'CanESM5', 'CAS-ESM2-0','MPI-ESM1-2-HR', 'MRI-ESM2-0', 'CESM2', 'GFDL-ESM2M']

rmadvect_terms_fafpas = rmadvect_temp_terms(model, 'faf-passiveheat')
```

```python
plt.figure(figsize=(28,20))
vmin = - 20
vmax = - vmin

for i in range(len(rmadvect_terms_fafpas)):
    plt.subplot(3, 3, i+1)
    if i == 4:
    # See level dimensions for MRI: they are slightly different to the other models
        rmadvect_terms_fafpas[i].mean('year').isel(lev=2).plot(vmin=vmin, vmax=vmax, cmap='bwr')
            
    else:
        rmadvect_terms_fafpas[i].mean('year').isel(lev=0).plot(vmin=vmin, vmax=vmax, cmap='bwr')
    plt.title(f'{model[i]}', fontsize=14)
```

### Examining T-z heat budget closure for ACCESS-CM2. 

```python
# Load in binned tendency terms

access_tend = xr.open_mfdataset('/g/data/e14/mp2135/FAFMIP_binned/ACCESS-CM2/faf-passiveheat/ocontemptend_binned.nc')['ocontemptend']              # Net time tendency due to all processes
access_rmadvect = xr.open_mfdataset('/g/data/e14/mp2135/FAFMIP_binned/ACCESS-CM2/faf-passiveheat/ocontemprmadvect_binned.nc')['ocontemprmadvect']  # Time tendency due to residual mean advection
access_pmdiff = xr.open_mfdataset('/g/data/e14/mp2135/FAFMIP_binned/ACCESS-CM2/faf-passiveheat/ocontemppmdiff_binned.nc')['ocontemppmdiff']        # Time tendency due to parameterised mesoscale diffusion
access_diff = xr.open_mfdataset('/g/data/e14/mp2135/FAFMIP_binned/ACCESS-CM2/faf-passiveheat/ocontempdiff_binned.nc')['ocontempdiff']              # Time tendency due to parameterised diapycnal diffusion + other vertical boundary layer processes (see OMIP Appendix L)
access_hfds = xr.open_mfdataset('/g/data/e14/mp2135/FAFMIP_binned/ACCESS-CM2/faf-passiveheat/hfds_binned.nc')['hfds']                              # Net surface heat flux
access_rsdoabsorb = xr.open_mfdataset('/g/data/e14/mp2135/FAFMIP_binned/ACCESS-CM2/faf-passiveheat/rsdoabsorb_binned.nc')['rsdoabsorb']            # Shortwave penetration
```

```python
# Examining closure of heat budget in T-z space for ACCESS-CM2. See L4.9 of Griffies et al. (2016) OMIP paper.
# Note: hfds is zero everywhere except the surface layer

LHS = access_tend
RHS = access_hfds + access_rsdoabsorb + access_rmadvect + access_pmdiff + access_diff

plt.figure(figsize=(28,10))
fontsize=14

year_index = 0

vmin = 1.55e13
vmax = -1.4e13
# For some reason having vmax = -vmin does not give a diverging colorbar...

plt.subplot(131)
LHS.isel(year=year_index).sel(lev=slice(200,None)).plot(yincrease=False, vmin=vmin, vmax=vmax, cmap='RdBu_r')
plt.title('LHS', fontsize=fontsize)

plt.subplot(132)
RHS.isel(year=year_index).sel(lev=slice(200,None)).plot(yincrease=False, vmin=vmin, vmax=vmax, cmap='RdBu_r')
plt.title('RHS', fontsize=fontsize)

plt.subplot(133)
(LHS - RHS).isel(year=year_index).sel(lev=slice(200,None)).plot(yincrease=False, vmin=vmin, vmax=vmax, cmap='RdBu_r')
plt.title('Residual', fontsize=fontsize)
```

The grid-cell heat budget is calculated based on that laid out in Griffies et al. (2016) (the OMIP paper) Section L4.9. 

For ACCESS-CM2, there is in most years a large blob of residual cooling at the surface for $10^\circ$C $< \Theta < 20^\circ$C. As Griffies et al. (2016) notes, the residual processes may include non-local KPP mixing, mixing from overflow schemes, and mixing across unresolved straits.

There is also some residual cooling for $\Theta < 0^\circ$C for $3000$ m $< z < 1000$ m, albeit smaller than the warmer surface residual.

### Temperature-depth streamfunctions in faf-passiveheat runs

```python
def fafpassive_TZstreamfuncs(models):
    
    streamfuncs = []
    
    for model in models:
        
        try:
            
            if model == 'ACCESS-CM2':
                rmadvect_binned = xr.open_mfdataset(f'/g/data/e14/mp2135/FAFMIP_binned/{model}/faf-passiveheat/ocontemprmadvect_binned.nc')['ocontemprmadvect']
                rmadvect_binned_streamfunc = tendency_term_binnned_streamfunction(rmadvect_binned, 'temperature')
                streamfuncs.append(rmadvect_binned_streamfunc)
                
            else:
                rmadvect_binned = xr.open_mfdataset(f'/g/data/e14/mp2135/FAFMIP_binned/{model}/faf-passiveheat/opottemprmadvect_binned.nc')['opottemprmadvect']
                rmadvect_binned_streamfunc = tendency_term_binnned_streamfunction(rmadvect_binned, 'temperature')
                streamfuncs.append(rmadvect_binned_streamfunc)
                
        except:
            
            streamfuncs.append(np.empty(0))
            print(f'{model}\'s rmadvect_binned not available for faf-passiveheat')
        
        print(f'{model} done')
        
    return streamfuncs
```

```python
models = ['ACCESS-CM2', 'CAS-ESM2-0', 'CanESM5', 'MIROC6', 'MPI-ESM1-2-HR', 'MRI-ESM2-0', 'CESM2', 'GFDL-ESM2M']

fafpassive_streamfuncs = fafpassive_TZstreamfuncs(models)
```

```python
fig = plt.figure(figsize=(25,20))

vmin = -20
vmax = -vmin

j=0

for i in range(len(fafpassive_streamfuncs)):
    j+=1
    try:
        ax = plt.subplot(3, 3, j)
        (fafpassive_streamfuncs[i]).isel(year=slice(-20,None)).mean('year')[:,:24].plot(yincrease=False, cmap='bwr', vmin=vmin, vmax=vmax, cbar_kwargs={"label": "Sv"})
        plt.title(f'{models[i]}', fontsize=14)
    except:
        print(f'{models[i]} not available')
        fig.delaxes(ax)
        j-=1
```
